All of these follow the same general idea as they are supposed to give a general description for a movie in some fashion. At this point, every movie in the dataset should have a decent amount of textual information describing it. If it isn't in the dataset at this point, the movie is so unknown that there is almost no information on it.

**Fixing Lists and Dictionaries**

```python
movies.info()
```

Unfortunately, by saving all the data in a CSV format, all the lists and dictionaries were converted to strings. So for each column that contains information in this format, it needs to be fixed. The columns that need fixed include:
- `spoken_languages`
- `genres`
- `keywords`
- `casts`
- `production_companies`

```python
for column in ['spoken_languages', 'genres', 'keywords', 'casts', 'production_companies']:
    movies[column] = movies[column].apply(lambda x: literal_eval(x))
```

With that being done, pickle the information (so it doesn't need evaluated again).

```python
filepath = dirpath + '.pkl'
if os.path.exists(filepath) != True:
    movies.to_pickle(filepath)
else:
    movies = pd.read_pickle(filepath)
```

Extract the most useful information regarding those lists and dictionaries.

```python
movies['spoken_languages'] = movies.spoken_languages.apply(lambda x: [i['name'] for i in x])
movies['genres'] = movies.genres.apply(lambda x: [i['name'] for i in x])
movies['keywords'] = movies.keywords.apply(lambda x: [i['name'] for i in x['keywords']])
movies['acting_top_5'] = movies.casts.apply(lambda x: [i['name'] for i in x['cast'] if i['order'] < 5])
movies['director'] = movies.casts.apply(lambda x: [i['name'] for i in x['crew'] if i['job'] == 'Director'])
movies['writers'] = movies.casts.apply(lambda x: list(set([i['name'] for i in x['crew'] if i['department'] == 'Writing'])))
movies['production_companies'] = movies.production_companies.apply(lambda x: [i['name'] for i in x])
```

The original title doesn't have much of a place here as any text that is being tokenized will require English as the language (which is what the `title` represents). The original title can always be queried using the API later on anyway if it happens to be different.

```python
movies = movies.drop('original_title', axis=1)
```

Also drop `casts` column as the most prominent members of movies have been extracted to other columns (Directors, Top 5 Actors/Actresses, and Writers).

```python
movies = movies.drop('casts', axis=1)
```

At this point all the textual information has been mostly dealt with. There is still much more to do, but the specifics of what changes will be made to the textual information will be related to the modeling techniques used later on. ## 3. Numerical Data Exploration (Finding Obscurity)

```python
movies.describe()
```

A few things have already been mentioned about these variables. And limitations have already been made concerning them.
- Movie runtimes have been limited to between 60 and 240 minutes
- Budget is likely only useful to find more well known movies (as their budgetary info is easier to find)
- For revenue the same conclusion can be drawn as the above
- The vote count is going to be a very useful feature in determining obscurity (less votes, less known)

**Budget and Revenue**

When reading the documentation on the [budget and revenue section of TMDb's "Movie Bible"](https://www.themoviedb.org/bible/movie#59f73b759251416e71000007), the information contained in the budget and revenue sections are supposed to be in USD and not adjusted for inflation. Now, that doesn't guarantee that the user input information is correct for all these movies, but in general that should be the case. So it's important to get rid of any extremely large and erroneous budgets and revenues. A [quick Google search](https://www.google.com/search?q=most+expensive+movie+ever+made) reveals that the most expensive movie ever made was *Pirates of the Caribbean: On Stranger Tides* at 378.5 Million USD. So any movies with a greater budget than that must be erroneous and will be removed.

```python
movies[movies.budget > movies[movies.title == "Pirates of the Caribbean: On Stranger Tides"].budget.values[0]]
```

The first movie above should probably be dropped: it has very little information and what information it does have appears to be erroneous. After a quick look for Neptune's Daughter from 1914, information couldn't be found on this particular iteration of the movie. It appears there may have been a remake in 1949 of the same movie, but the 1914 version will be dropped as no information could be found regarding it.

```python
movies = movies[~(movies.budget > movies[movies.title == "Pirates of the Caribbean: On Stranger Tides"].budget.values[0])]
```

```python
movies[movies.revenue > movies.revenue.max()*.7]
```

It looks like the maximum revenue is mostly accurate, the highest grossing film (before inflation) to date, at this point, is actually [*Avengers: Endgame*](https://en.wikipedia.org/wiki/List_of_highest-grossing_films#Highest-grossing_films), which happens to be one of the films that just wasn't picked up in the data collection process. Another one that looks to have been missed is *Titanic*. But these are very well known movies anyways, so they wouldn't get recommended in any case. A quick peek at the popularity scores above is a decent indicator for what should be expected for well known but not brand new movies. Next up, budgets and revenues will be adjusted for inflation based on the release year of the movie. Thus, movies with no release date information will be removed.

```python
movies = movies[~movies.release_date.isna()].copy()
```

```python
movies['release_year'] = movies.release_date.apply(lambda x: int(x.split('-')[0]))
```

```python
movies = movies[movies.release_year >= 1913].copy()
```

```python
def adj_budget_inflation(x):
    if x['release_year'] >= 2019:
        return x['budget']
    else:
        return round(inflate(x['budget'], x['release_year']))
def adj_revenue_inflation(x):
    if x['release_year'] >= 2019:
        return x['revenue']
    else:
        return round(inflate(x['revenue'], x['release_year']))
```

```python
movies['inf_adj_budget'] = movies.apply(adj_budget_inflation, axis=1)
movies['inf_adj_revenue'] = movies.apply(adj_revenue_inflation, axis=1)
```

```python
movies[movies.inf_adj_revenue > movies.inf_adj_revenue.max()*.8]
```

And as expected, *Gone with the Wind* has the highest revenues after inflation is taken into account, and nothing else really even comes close. In terms of this project, however, the budgets and revenues are more of qualifiers than actual useful indicators as they don't really add much useful information for this recommendation system and it is also hard to judge their reliability.

## 4. Into the Unknown (Making Sense of the Obscure)

```python
less_popular_movies.shape[0] / movies.shape[0]
```

The good news is, of the movies remaining in the movies dataset, the majority of movies in the set are mostly obscure, which means there is a lot of information. The bad news is, the more obscure the movie, the less confidence there is on the completeness of its information. This begs the question, are some movies are *too obscure*? For one thing, if its hard to find information about a movie from just searching it on the Internet, then it's also probably going to be difficult to find a way to watch the movie at all! Eventually, it would be great to detect and recommend movies that appear on streaming platforms so individuals using the service could go straight to a link to actually watch the movie. For the prototype phase, before it is even confirmed that this recommendation system even works very well, putting this infrastructure into place would be overkill. So for now, there won't be anything in place to check availability, but further down the line, given more iterations of this project, this would be essential to add for usability. To try and account for this, in the meantime, only movies with a minimum number of votes/ratings will be included in the final pool. When the models are trained, the entire set will be used (to get a good idea of the overall corpus). But for actual recommendations only movies that have some votes (and thus have actually been viewed by some users) will be included, assuming that these movies can be found and viewed in one form or another. 

```python
recommendation_pool = less_popular_movies[less_popular_movies.vote_count >= 20].copy()
```

```python
recommendation_pool.describe()
```

```python
plt.figure(figsize=(16,6))
sns.distplot(recommendation_pool.vote_average, bins=20)
plt.title('Vote Average (Obscure Movies) Distribution')
plt.xlabel('Vote Average')
plt.show()
```

Average scores for obscure films appear to follow a normal distribution.

```python
plt.figure(figsize=(16,6))
plt.scatter(recommendation_pool.popularity, recommendation_pool.vote_average, marker='o', alpha=.2)
plt.title('Vote Average vs. Popularity (Obscure)')
plt.xlabel('Popularity')
plt.ylabel('Vote Average')
plt.show()
```

It seems more popular does not mean better.

```python
plt.figure(figsize=(16,6))
sns.distplot(recommendation_pool.release_year, kde=False)
plt.title('Obscure Movies Release Years Distribution')
plt.xlabel('Release Year')
plt.show()
```

This makes sense for a number of reasons:
- More movies are being made because it is easier to make them with technological advancements
- More recent movies are more likely to have more ratings due to their recency and the rise of the Internet (everyone is able to be a movie reviewer/rater now and can share their opinions online)
- TMDb started out in 2008, so movies after that time naturally would have more votes (a new movie comes out, you see it, you rate it - likely not the case for older movies)

This is all useful to keep in mind moving forward.

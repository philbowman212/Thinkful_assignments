<h1 align='center'>Suicide Rates Dataset</h1>

```python
df = pd.read_csv('suicide_rates.csv')
```

```python
df.age.unique()
```

```python
df['age'] = df['age'].astype('category')
```

```python
df.age.cat.reorder_categories(['5-14 years','15-24 years','25-34 years','35-54 years','55-74 years','75+ years'], inplace=True)
```

## One Variable - Four Ways

```python
sns.set(style='whitegrid')
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

sns.barplot(y='suicides/100k pop',x='sex',hue='age',data=df, ax=axes[0])
axes[0].set_title('Bar by Sex and Age')
sns.barplot(y='suicides/100k pop',x='age',hue='sex',data=df, ax=axes[1])
axes[1].set_title('Bar by Age and Sex')
sns.pointplot(y='suicides/100k pop',x='sex',hue='age',data=df, join=False, dodge=True, ax=axes[2])
axes[2].set_title('Point by Sex and Age')
sns.pointplot(y='suicides/100k pop',x='age',hue='sex',data=df, join=False, dodge=True, ax=axes[3])
axes[3].set_title('Point by Age and Sex')

fig.suptitle('Suicides/100k Population Four Ways',y=1.01)
plt.tight_layout()
plt.show()
```

## Two Continuous - Three Ways

```python
sns.set(style='white')
fig, axs = plt.subplots(1, 3, figsize=(12,5))

sns.regplot(x='population', y='suicides_no', data=df, marker='.', fit_reg=False, scatter_kws={'s':2,'alpha':.5}, ax=axs[0])
axs[0].set_title('Scatterplot without Regression')

sns.kdeplot(df['population'], df['suicides_no'], cmap="Reds", shade=True, ax=axs[1])
axs[1].set_title('Heatmap')

sns.regplot(x='population', y='suicides_no', data=df, marker='.', scatter_kws={'s':2,'alpha':.5}, ax=axs[2])
axs[2].set_title('Scatterplot with Regression')

fig.suptitle('Compare Number of Suicides to Population', y=1.01)
plt.tight_layout()
plt.show()
```

## One Continuous, One Categorical - Six Ways

```python
df['generation'] = df['generation'].astype('category')
```

```python
df['generation'].cat.reorder_categories(['G.I. Generation', 'Silent', 'Boomers', 'Generation X', 'Millenials', 'Generation Z'],
                                        inplace=True)
```

```python
sns.set(style='ticks')
fig, axs = plt.subplots(3, 2, figsize=(12,15))
axs = axs.flatten()

sns.barplot(y='suicides/100k pop', x='generation', data=df, ax=axs[0])
axs[0].set_title('Overall Bar')

sns.pointplot(y='suicides/100k pop', x='generation', data=df, ax=axs[1])
axs[1].set_title('Overall Point')

sns.barplot(y='suicides/100k pop', x='generation', hue='sex', data=df, ax=axs[2])
axs[2].set_title('Bar Seperated by Sex')

sns.pointplot(y='suicides/100k pop', x='generation', hue='sex', data=df, ax=axs[3])
axs[3].set_title('Point Seperated by Sex')

sns.barplot(y='suicides/100k pop', x='generation', hue='age', data=df, ax=axs[4])
axs[4].set_title('Bar Seperated by Age')

sns.pointplot(y='suicides/100k pop', x='generation', hue='age', data=df, ax=axs[5])
axs[5].set_title('Point Seperated by Age')

fig.suptitle('Suicide Rates by Generation by Various Features', y=1.01)
plt.tight_layout()
plt.show()
```

<h1 align='center'>World Happiness Dataset</h1>

```python
wh = pd.read_csv('world_happiness_2016.csv')
```

```python
cols = list(wh.columns[6:])
```

```python
def quick_strip(column):
    plt.figure(figsize=(15,5))
    sns.stripplot(x='Region',y=column,data=wh)
    plt.xticks(rotation=30)
    plt.title('{} Score by Region'.format(column))
    plt.show()
```

```python
for col in cols:
    quick_strip(col)
```

```python
def comp_to_gdp(column):
    sns.regplot(wh.columns[6],column,wh)
    plt.show()
def comp_to_fam(column):
    sns.regplot(wh.columns[7],column,wh)
    plt.show()
def comp_to_health(column):
    sns.regplot(wh.columns[8],column,wh)
    plt.show()
def comp_to_freedom(column):
    sns.regplot(wh.columns[9],column,wh)
    plt.show()
def comp_to_trust(column):
    sns.regplot(wh.columns[10],column,wh)
    plt.show()
def comp_to_generosity(column):
    sns.regplot(wh.columns[11],column,wh)
    plt.show()
```

```python
for col in cols[1:]:
    comp_to_gdp(col)
for col in cols[2:]:
    comp_to_fam(col)
for col in cols[3:]:
    comp_to_health(col)
for col in cols[4:]:
    comp_to_freedom(col)
```
